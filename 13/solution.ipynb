```python
import copy
```

```python
# Write function to compare simple ints
def compare_integers(left, right):
    # If right order return 0
    if left < right:
        return 0
    # If right order return 1
    if left > right:
        return 1
    # Return 2 if nothing is decided
    return 2
```

# Task 1

```python
def compare(left, right, level=0):
    left = copy.deepcopy(left)
    right = copy.deepcopy(right)
    result = 2
    while (len(left) and len(right)) > 0:
        element_left = left.pop(0)
        element_right = right.pop(0)

        # Comparison if both are integers
        if isinstance(element_left, int) and isinstance(element_right, int):
            if compare_integers(element_left, element_right) == 0:
                return 0
            elif compare_integers(element_left, element_right) == 1:
                return 1
        
        # Comparison if both are lists
        if isinstance(element_left, list) and isinstance(element_right, list):
            intermediary_solution = compare(element_left, element_right, level=level+1)
            if intermediary_solution in [0, 1]:
                return intermediary_solution
        
        # Comparison if one is an integer and the other is a list
        if isinstance(element_left, list) and isinstance(element_right, int):
            element_right = [element_right]
            intermediary_solution = compare(element_left, element_right, level=level+1)
            if intermediary_solution in [0, 1]:
                return intermediary_solution
        if isinstance(element_left, int) and isinstance(element_right, list):
            element_left = [element_left]
            intermediary_solution = compare(element_left, element_right, level=level+1)
            if intermediary_solution in [0, 1]:
                return intermediary_solution

    # Comparison if one runs out of elements
    if len(left) > len(right):
        return 1
    elif len(left) < len(right):
        return 0
```

```python
# Read input and split into pairs
with open("input.txt") as file:
    pairs = []
    for i, pair in enumerate(file.read().split("\n\n")):
        clean_pair = []
        for package in pair.split("\n"):
            try:
                clean_pair.append(eval(package))
            except SyntaxError:
                continue
        pairs.append(clean_pair)
```

```python
results = []
for pair in pairs:
    left = pair[0]
    right = pair[1]
    results.append(compare(left, right))
```

```python
# Calculate the result
result = 0
for i, value in enumerate(results):
    if value == 0:
        result += i+1

result
```

# Task 2

```python
pairs_ordered = []
for i, result in enumerate(results):
    if result == 0:
        pairs_ordered.append(pairs[i][0])
        pairs_ordered.append(pairs[i][1])
    else:
        pairs_ordered.append(pairs[i][1])
        pairs_ordered.append(pairs[i][0])

len(pairs_ordered)
```

```python
divider_packet_first = [[2]]
divider_packet_second = [[6]]

pairs_ordered.append(divider_packet_first)
pairs_ordered.append(divider_packet_second)
```

```python
length = len(pairs_ordered)
over = length - 1
round_ = 1
while over > 0:
    for i in range(length-1):
        switch = compare(pairs_ordered[i], pairs_ordered[i+1])
        if switch == 0:
            over -= 1
            continue
        else:
            pairs_ordered[i], pairs_ordered[i+1] = pairs_ordered[i+1], pairs_ordered[i]
    print(f"{over} switches in round {round_}")
    if over == 0:
        break
    over = length - 1
    round_ += 1
```
